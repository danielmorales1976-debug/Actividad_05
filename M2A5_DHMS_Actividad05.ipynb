### a) Una sala de cine cobra diferentes precios de entrada según la edad de la persona. Si una persona es menor de 3 años, la entrada es gratuita; si tienen entre 3 y 12, el boleto es de 10 pesos; y si son mayores de 12 años, el boleto cuesta 15.00 pesos. Escribe un bucle en el que se solicite a los usuarios su edad y despliegue el costo de la entrada.

```python
while (edad := int(input("Indica la edad:"))) != 0 :
  if (edad < 3) :
    print(f"Tu edad es: {edad}.\n - Al ser menor de 3 años la entrada es gratuita")
  elif (edad <= 12) :
    print(f"Tu edad es: {edad}.\n - Al tener entre 3 y 12 años, tu boleto es de 10 pesos")
  else :
    print(f"Tu edad es: {edad}.\n - Tienes mas de 12 años, tu boleto cuesta 15 pesos")
```

### b) Crea un diccionario que contenga la información de una canción (artista, álbum, duración y género) y despliega el contenido de cada elemento (keys, values y tuplas)

```python
canciones = {
    "Bohemian Rhapsody": ["Queen","A Night at the Opera","5:55","Art Rock, Pop Rock"],
    "Shape of You": ["Ed Sheeran","Divide","3.53","Pop"],
    "Sound of Silence" : ["Simon and Garfunkel","Sounds of Silence","3:05","Folk pop"],
    "Sweet Caroline" : ["Neil Diamond","3:21","Soft rock"]
}
def mostrar_diccionario(diccionario):
    for clave, valor in diccionario.items():
        print(f"Clave: {clave}")
        print(f"Valor: {valor}")
        print("Tupla:", tuple(valor))
        print()
mostrar_diccionario(canciones)
```
